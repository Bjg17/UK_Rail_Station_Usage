MERGING THE TWO DATASET TO START ANALYSIS

```python
merged_data = pd.merge(data, previous_year_data, on = 'station', how = 'inner')
```

```python
merged_data.describe()
```

```python
merged_data['increase/decrease'] = (merged_data['entries_all_tickets_23/24'] - merged_data['entries_all_tickets_22/23'])
```

```python
merged_data['percentage_move'] = round(((merged_data['entries_all_tickets_23/24'] - merged_data['entries_all_tickets_22/23']) / merged_data['entries_all_tickets_22/23']),2)
```

```python
top_10_increase = merged_data.sort_values(by = 'percentage_move', ascending = False).head(10)
top_10_increase = top_10_increase[['station','region_23/24','entries_all_tickets_22/23','entries_all_tickets_23/24','increase/decrease','percentage_move']].reset_index()
top_10_increase.drop(columns = ['index'], inplace = True)
top_10_increase.rename(columns = {'region_23/24': 'region'}, inplace = True)
#top_10_increase.to_csv(r"C:\Users\banta\OneDrive\Desktop\real_project\top_10_increase.csv")

top_10_increase
```

```python
top_10_decrease = merged_data.sort_values(by = 'percentage_move').head(10)
top_10_decrease = top_10_decrease[['station','region_23/24','entries_all_tickets_22/23','entries_all_tickets_23/24','increase/decrease','percentage_move']].reset_index()
top_10_decrease.drop(columns = ['index'], inplace = True)
top_10_decrease.rename(columns = {'region_23/24': 'region'}, inplace = True)
#top_10_decrease.to_csv(r"C:\Users\banta\OneDrive\Desktop\real_project\top_10_decrease.csv")
top_10_decrease
```

```python
top_10_busiest_2223 = merged_data.sort_values(by = 'entries_all_tickets_22/23', ascending = False).head(10)

def list(num):
    num = int(num)
    if num >= 1_000_000_000_000:
        return f"{num/1_000_000_000: .2f}T"
    elif num >= 1_000_000_000:
        return f"{num/1_000_000 : .2f}B"
    elif num >= 1_000_000:
        return f"{num/1_000_000: .2f}M"
    elif num >= 1_000:
        return f"{num/1_000: .2f}K"
    else:
        return num
        
top_10_busiest_2223['total_entries/exits_22/23'] = top_10_busiest_2223['entries_all_tickets_22/23'].apply(list)

top_10_busiest_2223 = top_10_busiest_2223[['station','region_23/24', 'entries_all_tickets_22/23','total_entries/exits_22/23']].reset_index()
top_10_busiest_2223.drop(columns = ['index'], inplace = True)
top_10_busiest_2223.rename(columns = {'region_23/24': 'region'}, inplace = True)
top_10_busiest_2223


#top_10_busiest_2223.to_csv(r"C:\Users\banta\OneDrive\Desktop\real_project\top_10_busiest_2223.csv")
top_10_busiest_2223
```

```python
top_10_busiest_2324 = merged_data.sort_values(by = 'entries_all_tickets_23/24', ascending = False).head(10)
top_10_busiest_2324['total_entries/exits_23/24'] = top_10_busiest_2324['entries_all_tickets_23/24'].apply(list)

top_10_busiest_2324 = top_10_busiest_2324[['station','region_23/24','entries_all_tickets_23/24','total_entries/exits_23/24']].reset_index()
top_10_busiest_2324.drop(columns = ['index'], inplace = True)
top_10_busiest_2324.rename(columns = {'region_23/24': 'region'}, inplace = True)
#top_10_busiest_2324.to_csv(r"C:\Users\banta\OneDrive\Desktop\real_project\top_10_busiest_2324.csv")

top_10_busiest_2324
```

```python
top_10_interchanges = merged_data.sort_values(by = 'interchanges_23/24', ascending = False).head(10)
top_10_interchanges['interchange_change_over_year'] = top_10_interchanges['interchanges_23/24']-top_10_interchanges['interchanges_22/23']
top_10_interchanges['interchange_percentage_change_over_year'] = round(((top_10_interchanges['interchanges_23/24']-top_10_interchanges['interchanges_22/23'])
                                                                        /top_10_interchanges['interchanges_22/23'].replace({0 : np.nan})),2)

top_10_interchanges['total_intercganges_22/23'] = top_10_interchanges['interchanges_22/23'].apply(list)
top_10_interchanges['total_intercganges_23/24'] = top_10_interchanges['interchanges_23/24'].apply(list)
top_10_interchanges['total_diff'] = top_10_interchanges['interchange_change_over_year'].apply(list)

top_10_interchanges = top_10_interchanges[['station','region_23/24','interchanges_22/23','total_intercganges_22/23', 
                                           'interchanges_23/24', 'total_intercganges_23/24','interchange_change_over_year',
                                          'interchange_percentage_change_over_year']].reset_index()
top_10_interchanges.drop(columns = ['index'], inplace = True)
top_10_interchanges.rename(columns = {'region_23/24': 'region'}, inplace = True)
top_10_interchanges.to_csv(r"C:\Users\banta\OneDrive\Desktop\real_project\top_10_interchanges.csv")

top_10_interchanges
```

```python
region_growth = merged_data.groupby('region_23/24')[['entries_all_tickets_22/23','entries_all_tickets_23/24']].sum()

region_growth['percentage_move'] = ((region_growth['entries_all_tickets_23/24'] - region_growth['entries_all_tickets_22/23'])
                                     / region_growth['entries_all_tickets_22/23'].replace({0 : np.nan})).round(2)
region_growth.reset_index(inplace = True)
region_growth.rename(columns={'region_23/24':'regions'}, inplace = True)
region_growth.sort_values(by = 'percentage_move', ascending = False, inplace = True)
region_growth.to_csv(r"C:\Users\banta\OneDrive\Desktop\real_project\region_growth.csv")

region_growth
```
